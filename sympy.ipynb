```python
import sympy as sp
x=sp.symbols("x")
y=sp.symbols("y")
f=(sp.sin(x) + x**2) * sp.exp(x*2) + sp.log((x + 1)/(x + 2))
```

```python
sp.integrate(f,(x,0,10))
```

```python
sp.diff(f)
```

```python
sp.limit(f,x,10)
```

```python
sp.plot(f, (x, -10, 10))
```
